```python
import pickle
import dash
from dash.dependencies import Input, Output, State
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import CountVectorizer
import dash_core_components as dcc
```

```python
app = dash.Dash()
```

```python
project_name = None
```

```python
def load_model():
    global df
    df = pd.read_csv('balanced_reviews.csv')
  
    global pickle_model
    file = open("pickle_model.pkl", 'rb') 
    pickle_model = pickle.load(file)

    global vocab
    file = open("feature.pkl", 'rb') 
    vocab = pickle.load(file)
    
    print(df.sample(5))
```

```python
def check_review(reviewText):
    #reviewText has to be vectorised, that vectorizer is not saved yet
    #load the vectorize and call transform and then pass that to model preidctor
    #load it later

    transformer = TfidfTransformer()
    loaded_vec = CountVectorizer(decode_error="replace",vocabulary=vocab)
    reviewText = transformer.fit_transform(loaded_vec.fit_transform([reviewText]))


    # Add code to test the sentiment of using both the model
    # [0] == negative   [1] == positive
    return pickle_model.predict(reviewText)
```

```python
def open_browser():
    webbrowser.open_new('http://127.0.0.1:8050/')
```

```python
df1 = pd.read_csv('final_reviews_for_ear_rings1')
df1['positivity'] = df1['reviews'].apply(lambda i:check_review(i)[0])
count1 = df1['positivity'].value_counts()
print(count1[1])
count1
```

```python
def create_app_ui():

    main_layout = html.Div([
        html.H1(children='Sentiments Analysis with Insights', id='Main_title',style={'text-align':'center','fontSize':50}),
        html.Div([
            html.Img(src = app.get_asset_url('review.png'))
        ], style={'textAlign': 'center'}),
        html.Div(style={'backgroundColor':'#52b7e9'},children=[
            html.H3(children='Pie Chart of reviews', id='sub_heading1',style={'color':'red','fontSize':30,'text-align':'center'}),
            dcc.Graph(figure=px.pie(df1,values=[count1[1],count1[0]],names=['positive_reviews','negative reviews'],hole=.4,opacity=1))
            #dcc.Graph(figure=px.histogram(df1,x='positivity', color=['positive_reviews','negative reviews'], marginal="rug"))
        ], className='piegraph_contents'),
        html.Div(style={'backgroundColor':'#111111'}, children=[
        html.Div([
            html.H3(children='Check reviews using dropdown list', id='sub_heading2',style={'color':'#7FDBFF','fontSize':20,'text-align':'center'}),

            dcc.Dropdown(
                id='dropdown',
                options=[{'label':i,'value':i} for i in df1["reviews"][:40]],
                value='I loved the shoes'

            ),
            html.Br(),
            html.Button(children='Find Review', id='button_review',n_clicks=0,style={'backgroundColor':'yellow','fontSize':20,'width':'10%','height':'40px'}),
            html.H1(children=None, id='result',style={'color':'#efcccf','text-align':'center'}),
        ], className='Dropdown_list'),
        html.Div([
            html.H3(children='Enter review in textbox and check', id='sub_heading3',style={'color':'#7FDBFF','fontSize':20,'text-align':'center'}),
            
            dcc.Textarea(
                
                id='textarea_review',
                placeholder='Enter the review here...',
                #value='My daughter loves these shoes',
                style={'width': '100%', 'height': 70,'backgroundColor':'#99ff99'}
                ),
            html.Br(),
            html.Button(children='Find Review',id='button_review1',n_clicks=0,style={'backgroundColor':'yellow','fontSize':20,'width':'10%','height':'40px'}),
            html.H1(children=None, id='result1',style={'color':'#efcccf','text-align':'center'}),
            
        ], className='text_review')
        ])
    ], className='main_window')
    
    return main_layout
```

```python
@app.callback(
    Output('result', 'children'),
    [
    Input('button_review', 'n_clicks'),
    State('dropdown','value')
    ]
    )
def update_app_ui(n_click,textarea_value):
    
    print("Data Type  = ", str(type(textarea_value)))
    print("Value      = ", str(textarea_value))

    
    result_list = check_review(textarea_value)
    if n_click > 0:
        if (result_list[0] == 0 ):
            result = 'Review is : Negative'
        elif (result_list[0] == 1 ):
            result = 'Review is : Positive'
        else:
            result = 'Review is : Unknown'
        
    return result
@app.callback(
    Output('result1', 'children'),
    [
    Input('button_review1','n_clicks'),
    Input('textarea_review', 'value')
    ]
    )
def update_app_ui1(n_clicks,textarea_value):
    
    print("Data Type  = ", str(type(textarea_value)))
    print("Value      = ", str(textarea_value))
    result_list = check_review(textarea_value)

    if n_clicks>0:
        print(result_list)
        if (result_list[0] == 0 ):
            result = 'Review is : Negative'
        elif (result_list[0] == 1 ):
            result = 'Review is : Positive'
        else:
            result = 'Review is : Unknown'

    return result
```

```python
# Main Function to control the Flow of your Project
def main():
    load_model()    
    open_browser()
    

    global project_name
    project_name = "Sentiments Analysis with Insights" 
      
    global app
    app.layout = create_app_ui()
    app.title = project_name
    # go to https://www.favicon.cc/ and download the ico file and store in assets directory 
    app.run_server() # debug=True
  
    print("This would be executed only after the script is closed")
    app = None
    project_name = None
```

```python
if __name__ == '__main__':
    main()
```
